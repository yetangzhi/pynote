```python
def apid2index(df,apid):
    return df[df['apid']==apid].index.values[0]
```

```python
ldamatrix = np.ndarray(shape=(len(macGroups),len(aps)), dtype=int)
```

```python
ldamatrix.shape
```

```python
def mac2index(df,mac):
    return df[df[0]== mac].index.values[0]
```

```python
macList= []
for eachmac,logs in macGroups:
    macList.append(eachmac)
```

```python
macDF = pd.DataFrame(macList)
```

```python
macDF.head()
```

```python
mac2index(macDF, '000f534b8f' )
```

```python
for eachMac, logs in macGroups:
    for eachlog in logs['apid']:
        ldamatrix[mac2index(macDF,eachMac)][apid2index(aps,eachlog)] += 1
```
