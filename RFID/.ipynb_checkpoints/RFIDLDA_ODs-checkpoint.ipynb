```python
import pandas as pd
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import math
import datetime
import splitTraj
import lda
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
```

```python
records = splitTraj.splitByNmins(datetime.datetime(2012,12,1),30)
```

```python
records.head()
```

```python
segmentDF = splitTraj.getSegmentDF(records)
```

```python
segmentDF.head()
```

```python
top2000word= segmentDF.word.value_counts().index[:2000].values.tolist()
```

```python
segmentDF_inTopWord = segmentDF[segmentDF['word'].isin(top2000word)]
```

```python
len(segmentDF_inTopWord)
```

```python
ldamatrix = pd.crosstab(index = segmentDF_inTopWord['MOVER_ID'],columns = segmentDF_inTopWord['word'])
```

```python
car_list = ldamatrix.index.values.astype(str).tolist()
od_list = ldamatrix.columns.values.astype(str).tolist()
```

```python
model = lda.LDA(n_topics=20, n_iter=1000, random_state=1)
```

```python
model.fit(ldamatrix.values.astype(int))
```

```python
topic_word = model.topic_word_
for i, topic_dist in enumerate(topic_word):
    topic_words =np.array(od_list)[np.argsort(topic_dist)][: -9:-1]
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))
```

```python
lng_min = 118.437226706165 #117.88333 
lng_max = 119.107761430939 #118.433333 
lat_min = 31.854224661961 #24.383333
lat_max = 32.145074111061 #24.9

offset = np.maximum(lng_max-lng_min,lat_max-lat_min)*0.1

map_extent = [lng_min-offset, lng_max+offset, lat_min-offset, lat_max+offset]
```

```python
map_tiles = cimgt.MapQuestOSM() # 载入MapQuest背景地图
ax = plt.axes(projection=map_tiles.crs) # 设定地图投影

fig = ax.figure
fig.set_size_inches(16,16) # 设定地图绘制大小

ax.set_extent(map_extent) # 设定地图空间范围
ax.add_image(map_tiles, 11, alpha=0.5) # 添加背景地图，并设置缩放级别为11级，地图透明度为0.5

#ax.plot(mergeWay['lon'], mergeWay['lat'], color='black', transform=ccrs.Geodetic()) # 将轨迹的经纬度位置绘制为地图上黑色的折线

ax.set_title('All Trajectories in One Day',fontsize=15) # 设定地图标题，并设置字号为15号

plt.tight_layout() # 使用紧凑布局
```
