```python
from lxml import etree 
from lxml import objectify
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib as plt
%matplotlib inline
```

```python
OSMfile = objectify.parse("xiamenOSM")
```

```python
root = OSMfile.getroot()
```

```python
print(root.tag)
```

```python
nID = []
nLat = []
nLon = []
nodes = []
for node in root.iter('node'):
     nID.append(int(node.get('id')))
     nLat.append(float(node.get('lat')))
     nLon.append(float(node.get('lon')))
     nodes.append(node)
```

```python
d = {'nid' : nID,'lat' : nLat,'lon' : nLon}
```

```python
dfNodes = pd.DataFrame(d)
```

```python
np.max(dfNodes)
```

```python
dfNodes.head()
```

```python
len(nodes)
```

```python
nodes[0].getchildren()[0].tag
```

```python
nodes[0].get("lat") 
```

```python
nodes[0].keys()
```

```python
ways = []
wid = []
wNid = []
wSeq = []
for way in root.iter('way'):
    ways.append(way)
    count = 0
    for nd in way.iter('nd'):
        wid.append(int(way.get("id")))
        wNid.append(int(nd.get("ref")))
        wSeq.append(count)
        count = count + 1
```

```python
len(ways)
```

```python
len(wid)
```

```python
dw = {"wid": wid,"nid" : wNid,"seq" : wSeq}
```

```python
dfWays = pd.DataFrame(dw)
```

```python
dfWays.head()
```

```python
mergeWay = pd.merge(dfWays,dfNodes)
```

```python
mergeWay = mergeWay.sort_index(by=['wid','seq'])
```
