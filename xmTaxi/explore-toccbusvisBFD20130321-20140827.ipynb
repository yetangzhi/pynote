# 公交车轨迹可视化示例

以下代码对公交车GPS轨迹数据进行最简单的可视化。代码将读入一辆公交车一天的轨迹数据，然后展示：

  - 一天的所有轨迹数据
  - 一天的所有上行轨迹数据
  - 一次上行的轨迹数据
    
王祖超 2014-08-27

### 载入所需的Python库

```python
from __future__ import print_function,division
import numpy as np
import pandas as pd
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
```

让绘图嵌入ipython notebook，而不是弹出一个新窗口

```python
%matplotlib inline
```

### 读入公交车GPS轨迹数据

读入csv数据文件，注意设置编码为utf-8，轨迹数据保存在df中

```python
df = pd.read_csv('../data_test/BFD20130321_G43410.csv',header=None,parse_dates=[5],encoding='utf-8')
```

给每一列取名字

```python
df.columns = ['company','id','plate_no','route','updown','time','longitude','latitude','velocity','direction','next_station_id']
```

对采样点按照时间排序

```python
df.sort(['time'],inplace=True)
```

```python
df.head()
```

```python
print(u'该文件总共包含%d条记录'%len(df))
```

确定数据的空间范围

```python
# 得到经纬度的范围
lng_min = np.min(df['longitude'])
lng_max = np.max(df['longitude'])
lat_min = np.min(df['latitude'])
lat_max = np.max(df['latitude'])

offset = np.maximum(lng_max-lng_min,lat_max-lat_min)*0.1

map_extent = [lng_min-offset, lng_max+offset, lat_min-offset, lat_max+offset]
```

### 展示一天的所有轨迹数据

```python
map_tiles = cimgt.MapQuestOSM() # 载入MapQuest背景地图
ax = plt.axes(projection=map_tiles.crs) # 设定地图投影

fig = ax.figure
fig.set_size_inches(8,8) # 设定地图绘制大小

ax.set_extent(map_extent) # 设定地图空间范围
ax.add_image(map_tiles, 11, alpha=0.5) # 添加背景地图，并设置缩放级别为11级，地图透明度为0.5

ax.plot(df['longitude'], df['latitude'], color='black', transform=ccrs.Geodetic()) # 将轨迹的经纬度位置绘制为地图上黑色的折线

ax.set_title('All Trajectories in One Day',fontsize=15) # 设定地图标题，并设置字号为15号

plt.tight_layout() # 使用紧凑布局
```

### 展示一天的上行轨迹

选择所有上行的轨迹数据，保存在dfup中

```python
dfup = df[df['updown']==1]
```

```python
print(u'一天的上行轨迹有%d个采样点'%len(dfup))
```

绘制dfup中的轨迹数据

```python
map_tiles = cimgt.MapQuestOSM()
ax = plt.axes(projection=map_tiles.crs)
fig = ax.figure
fig.set_size_inches(8,8)

ax.set_extent(map_extent)
ax.add_image(map_tiles, 11, alpha=0.5)

# 将轨迹的经纬度位置绘制为地图上的散点图
points = ax.scatter(dfup['longitude'], dfup['latitude'],
                    lw=0, # 散点无边界
                    s=20, # 散点大小为20
                    c=dfup['next_station_id'], # 散点颜色编码next_station_id
                    cmap = plt.get_cmap('jet'), # 使用彩虹色color scale
                    transform=ccrs.Geodetic())

ax.set_title('All Upstream Trajectories in One Day',fontsize=15)

plt.colorbar(points) # 添加color scale的legend

plt.tight_layout()
```

### 展示一次的上行轨迹

选择轨迹的开始点，也就是next_station_id减小的地方

```python
tripStIndex = np.arange(len(dfup))[np.concatenate(([False],dfup['next_station_id'].iloc[:-1]>dfup['next_station_id'].iloc[1:]))]
print(u'轨迹开始点编号：%s'%(','.join(tripStIndex.astype(unicode))))
```

因此这里有5条上行轨迹

选择从编号371到715之间的那一条上行轨迹，保存在dfup_onetrip中

```python
dfup_onetrip = dfup.iloc[tripStIndex[1]:tripStIndex[2]]
dfup_onetrip.head()
```

```python
dfup_onetrip.tail()
```

```python
print(u'这一条上行轨迹有%d个采样点',len(dfup_onetrip))
```

绘制dfup_onetrip中的轨迹数据

```python
map_tiles = cimgt.MapQuestOSM()
ax = plt.axes(projection=map_tiles.crs)
fig = ax.figure
fig.set_size_inches(8,8)

ax.set_extent(map_extent)
ax.add_image(map_tiles, 11, alpha=0.5)

points = ax.scatter(dfup_onetrip['longitude'], dfup_onetrip['latitude'],
                    lw=0,
                    s=20,
                    c=dfup_onetrip['next_station_id'],
                    cmap = plt.get_cmap('jet'),
                    transform=ccrs.Geodetic())

ax.set_title('One Upstream Trajectory',fontsize=15)

plt.colorbar(points)

plt.tight_layout()
```
