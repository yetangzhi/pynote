```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
%matplotlib inline
```

```python
record = pd.read_csv("../../data/xiamenTraj/2014_09_01_10/preClean/2014_09_01.csv")
```

```python
record.head()
```

```python
def printCarTraj(recordDF,MOVER_ID):
    carDF = recordDF[recordDF.MOVER_ID == MOVER_ID]
    lng_min = np.min(carDF.X)
    lng_max = np.max(carDF.X)
    lat_min = np.min(carDF.Y)
    lat_max = np.max(carDF.Y)
    offset = np.maximum(lng_max-lng_min,lat_max-lat_min)*0.1
    map_extent = [lng_min-offset, lng_max+offset, lat_min-offset, lat_max+offset]
    
    map_tiles = cimgt.MapQuestOSM() # 载入MapQuest背景地图
    ax = plt.axes(projection=map_tiles.crs) # 设定地图投影

    fig = ax.figure
    fig.set_size_inches(8,8) # 设定地图绘制大小


    ax.set_extent(map_extent) # 设定地图空间范围
    ax.add_image(map_tiles, 11, alpha=0.5) # 添加背景地图，并设置缩放级别为11级，地图透明度为0.5
    ax.plot(carDF.X,carDF.Y, transform=ccrs.Geodetic())
    ax.set_title("car : " + str(MOVER_ID))
    plt.tight_layout() # 使用紧凑布局
```

```python
def printCarCompareSpeed0(recordDF,MOVER_ID):
    carDF = recordDF[recordDF.MOVER_ID == MOVER_ID]
    lng_min = np.min(carDF.X)
    lng_max = np.max(carDF.X)
    lat_min = np.min(carDF.Y)
    lat_max = np.max(carDF.Y)
    offset = np.maximum(lng_max-lng_min,lat_max-lat_min)*0.1
    map_extent = [lng_min-offset, lng_max+offset, lat_min-offset, lat_max+offset]
    
    fig = plt.figure(figsize=(8,16))
    map_tiles = cimgt.MapQuestOSM() # 载入MapQuest背景地图
    ax = plt.subplot(121,projection=map_tiles.crs) # 设定地图投影
    ax.set_extent(map_extent) # 设定地图空间范围
    ax.add_image(map_tiles, 11, alpha=0.5) # 添加背景地图，并设置缩放级别为11级，地图透明度为0.5
    ax.plot(carDF.X,carDF.Y, transform=ccrs.Geodetic())
    ax.set_title("raw data of car : " + str(MOVER_ID))
    
    carDF = carDF[carDF.SPEED != 0]
    ax = plt.subplot(122,projection=map_tiles.crs) # 设定地图投影
    ax.set_extent(map_extent) # 设定地图空间范围
    ax.add_image(map_tiles, 11, alpha=0.5) # 添加背景地图，并设置缩放级别为11级，地图透明度为0.5
    ax.plot(carDF.X,carDF.Y, transform=ccrs.Geodetic())
    ax.set_title("remove speed 0 of car : " + str(MOVER_ID))
    plt.tight_layout() # 使用紧凑布局
```
