```python
jjq = []
with open('20141001_jjq.csv') as jjqf:
    jjqf.readline()
    for each in jjqf:
        record = each.strip().split(',')
        jjq.append(record)
```

```python
jjq_simset = set()
for each_record in jjq:
    jjq_simset.add(each_record[15])
```

```python
count = 0
for each_gps in gps:
    if(each_gps[0] in jjq_simset):
        count = count + 1
print(str (count/len(gps)) + " of gps in jjq Sim number set")
```
